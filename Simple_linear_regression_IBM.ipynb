# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Importing the dataset

```python
# Importing the dataset
dataset = pd.read_csv('/kaggle/input/salary/Salary.csv')

# Separating the features (Years of Experience) and the target variable (Salary)
X = dataset.iloc[:, :-1].values  # Assuming the features are in all columns except the last one
Y = dataset.iloc[:, -1].values   # Assuming the target variable is in the last column
```

# Splitting the Dataset into Training and Test Sets

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

# Training the Simple Linear Regression model on the Training set

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

# Predicting the Test set results

```python
Y_pred = regressor.predict(X_test)
```

# Visualising the Training set results

```python
plt.scatter(X_train, Y_train, color='blue')  # Plotting the actual data points
plt.plot(X_train, regressor.predict(X_train), color='red')  # Plotting the regression line
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Visualising the Test set results

```python
plt.scatter(X_test, Y_test, color='blue')  # Plotting the actual data points
plt.plot(X_train, regressor.predict(X_train), color='red')  # The regression line is the same as for the training set
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Predicting the salary of an employee with a specific number of years of experience

```python
years = input("Enter the number of years of Experience: ")
years_float = float(years)  
prediction = regressor.predict([[years_float]]) 

print(f"The salary of an employee with {years_float} years of experience is {prediction[0]}")  # Displaying the prediction
```
